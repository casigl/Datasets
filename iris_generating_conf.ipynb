```
import pandas as pd
import numpy as np
from sklearn import linear_model
from pandas import read_csv
from sklearn import preprocessing

#Importing IRIS dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
iris_df = read_csv(url, header= None)
iris_df.columns = ['sepal_length', 'sepal_width', 'petal_length','petal_width','class']

# Create x, where x the 'scores' column's values as floats
x = iris_df[['sepal_length', 'sepal_width', 'petal_length','petal_width']]

normalized_X = pd.DataFrame(preprocessing.normalize(x))

normalized_X.columns = ['sepal_length', 'sepal_width', 'petal_length','petal_width']
print(normalized_X)
```

```
def compute_normal_diff(target, atr1, atr2, atr3):
  """ 
  Computes the absolute value difference between the LR model predicted target 
  value and the actual target value and returns the normalized error. Produces a 
  list of 150 errors for the target attribute. 

  Parameters
  ----------
    target: series
      series of normalized target attribute values
    attr1, attr2, attr3: series
      other series of normalized attribute values used in the LR algoithm

  Returns
  -------
    error_list: list
      contains all error displayed as confidence values for target attribute
  """
  total_difference = 0
  error_list = []

  #loop through all 150 examples in IRIS dataset
  for i in range (150):

    actual_target = target[i]
    
    predicted_target = reg.predict([[atr1[i], atr2[i], atr3[i]]])
    predicted_target = predicted_target.item() #converts 1-item array into float

    error = 0
    difference = 0
    difference = abs(predicted_target - actual_target)


    if (difference != 0):
      #give preference to dividing by the actual target
      if (difference < actual_target):
        error = (difference/actual_target)
      #else divide by predicted
      #only goes into the else 4 times for petal width
      else:
        error = (difference/predicted_target)

    confidence = (1-error)*100

    error_list.append(confidence)

  return error_list
```
